```python
import os
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def zero_count(li):
    count=-1
    for i in li:
        if i==0:
            count=count+1
    return count
            
```

```python
dir_name = 's0.2/Files/Time Rate Data/'
li = os.listdir(dir_name)
print(li)
output_file = open("graph_results/s.2.txt","a")



nor_y_list = []
y_list = []
step = 100
rc = 1
f_no=1
index=0
nor_sum=0
sum=0
for f_name in li:
    
    array1 = []
    file_name = dir_name + f_name
    with open(file_name) as file:
        reader = csv.reader(file, delimiter=' ')
        for row in reader:
            array1.append(int(row[1]))
        result = zero_count(array1)

        nor_sum=nor_sum+(result/len(array1))
        sum = sum + result
        if rc==100:
            rc=1
            nor_y_list.append(nor_sum/100)
            y_list.append(sum/100)
            sum=0
            nor_sum=0
        rc = rc+1
        
    

        output_file.write(str(result) +" "+ str(len(array1)) +" "+ str(result/len(array1)) +"\n")
output_file.close()
```
